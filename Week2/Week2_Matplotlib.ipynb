```python
import matplotlib.pyplot as plt
import numpy as np

# data
# [0, 0.1, 0.2, 0.3, ... , 9.9]
x = np.arange(0,10,0.1)
y= np.sin(x)

# graph
plt.plot(x, y)

# determine chart properties
plt.title("sinus graph")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = np.random.randint(50,100,50)

plt.scatter(x, y, c = colors, s = sizes)

plt.title("Dot graph")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```python
# data
data = np.random.randn(100000)

plt.hist(data, bins=30)

plt.title("histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#data
x = ['A', 'B', 'C', 'D', 'E']
y = np.random.randint(1,100,5)

plt.bar(x,y)

plt.title("Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```
